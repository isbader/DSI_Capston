```python
w = Movie_names['cast']
for c in range (len(w)):
    print(c,type(w[c]))
    
```

```python

castnames=[]
for x in w:
    thisn = []
    try:
        l = x.split(',')
        for n in l:
            rindx = n.find('name:_')
            endindx = n.find('_>',rindx)
            mem = n[rindx+6:endindx]
            thisn.append(mem)
        castnames.append(thisn)
    except:
        castnames.append(None)
```

```python
len(castnames)
```

```python
Movie_names['cast'] = castnames
```
