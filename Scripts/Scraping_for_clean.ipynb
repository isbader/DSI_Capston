```python
movies = pd.read_csv('/Users/bader/Dropbox/DSI_GA_Capstone/Code/New_Data_OMDB/movies_2000_new.csv')
```

```python
movies['imdbID']
```

```python
ID = movies['imdbID'][0:10]
name = []
runtime = []
genres = []
plot = []

for i in ID:
    try:
        if type(i) == str:
            link = 'https://pro.imdb.com/title/{}/details'.format(i)
            print(link)
            response = requests.get(link)
            print(response.status_code)
            soup = BeautifulSoup(response.text, 'html.parser')
            try:
                name.append(soup.find('h1').text)
            except:
                name.append(None)
            try:
                runtime.append(soup.select('span[id^="running_time"]')[0].text)
            except:
                runtime.append(None)
            try:
                genres.append(soup.select('span[id^="genres"]'))
            except:
                genres.append(None)
            try:
                plot.append(soup.select('div[id^="title_summary"]')[0].text.split('\n')[1])
            except:
                plot.append(None)
                
    except:
        name.append(None)
        runtime.append(None)
```

```python
print(len(ID))
print(len(name))
print(len(runtime))
print(len(genres))
print(len(plot))
```

```python
ID_ = ID.dropna()
```

```python
print(len(ID_))
print(len(name))
print(len(runtime))
print(len(genres))
print(len(plot))
```

```python
df_ = pd.DataFrame({'ID':ID_,'name':name,'runtime':runtime,'genres':genres,'plot':plot})
```
