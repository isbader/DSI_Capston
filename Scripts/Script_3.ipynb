```python
Movie_names = pd.read_csv('names/names_2009_.csv') #Get the movie names
Movies_2016_2017 = Movie_names 
Movies_2016_2017.drop_duplicates(keep='first',inplace=True) 
```

```python
runtime = []
cast = []
genres = []
runtimes = []
country_codes = []
language_codes = []
certificates = []
box_office = []
rating = []
votes = []
original_air_date = []
title = []
directors = []
writers = []
producers = []
composers = []
editors = []
cinematographers = []
set_decorators = []
production_companies = []
assistant_directors = []
visual_effects = []

count = 0
for i in Movies_2016_2017['Movie_ID']:
    
    movie = moviesDB.get_movie(i)
    
    try:
        cast.append(movie['cast'])
    except KeyError:
        cast.append(None)
        
    try:
        runtime.append(movie['runtime'][0])
    except KeyError:
        runtime.append(None)
    
    try:
        genres.append(movie['genres'])
    except KeyError:
        genres.append(None)
        
    try:
        country_codes.append(movie['country codes'])
    except KeyError:
        country_codes.append(None)
        
    try:
        language_codes.append(movie['language codes'])
    except KeyError:
        language_codes.append(None)
    
    try:
        USRate = [s for s in movie['certificates'] if "United States" in s]
        if len(USRate) > 0:
            certificates.append(USRate[0].replace('United States:',''))
        else:
            certificates.append(None)

    except KeyError:
        certificates.append(None)
        
    try:
        box_office.append(movie['box office'])
    except KeyError:
        box_office.append(None)
        
    try:
        rating.append(movie['rating'])
    except KeyError:
        rating.append(None)
        
    try:
        votes.append(movie['votes'])
    except KeyError:
        votes.append(None)
    
    try:
        original_air_date.append(movie['original air date'])
    except KeyError:
        original_air_date.append(None)
        
    try:
        title.append(movie['title'])
    except KeyError:
        title.append(None)
        
    try:
        directors.append(movie['directors'])
    except KeyError:
        directors.append(None)
        
    try:
        writers.append(movie['writers'])
    except KeyError:
        writers.append(None)
        
    try:
        producers.append(movie['producers'])
    except KeyError:
        producers.append(None)
        
    try:
        composers.append(movie['composers'])
    except KeyError:
        composers.append(None)
        
    try:
        editors.append(movie['editors'])
    except KeyError:
        editors.append(None)
        
    try:
        cinematographers.append(movie['cinematographers'])
    except KeyError:
        cinematographers.append(None)
        
    try:
        set_decorators.append(movie['set decorators'])
    except KeyError:
        set_decorators.append(None)
        
    try:
        production_companies.append(movie['production companies'])
    except KeyError:
        production_companies.append(None)
        
    try:
        assistant_directors.append(movie['assistant directors'])
    except KeyError:
        assistant_directors.append(None)
        
    try:
        visual_effects.append(movie['visual effects'])
    except KeyError:
        visual_effects.append(None) 
    count= count +1
    print(count,i)
```

```python
# cast 

cast_=[]

for d in cast:
    try:
        cast_.append(str(d[0]))

    except:
        cast_.append(None)


# directors 

director=[]

for d in directors:
    try:
        director.append(str(d[0]))

    except:
            director.append(None)


#producers

producer1=[]
producer2=[]

for d in producers:
    try:
        producer1.append(str(d[0]))
        if len(d)>1:
            producer2.append(str(d[1]))
        else:
            producer2.append(None)
    except:
            producer1.append(None)
            producer2.append(None)

#writers

writer1=[]
writer2=[]
print ('writers',len(writers))
for d in writers:
    try:
        writer1.append(str(d[0]))
        if len(d)>1:
            writer2.append(str(d[1]))
        else:
            writer2.append(None)
    except:
            writer1.append(None)
            writer2.append(None)

        
        
#production_companies

production_company1 = []
production_company2 = []
print ('production_companies',len(production_companies))


for d in production_companies:
    try:
        production_company1.append(str(d[0]))
        if len(d)>1:
            production_company2.append(str(d[1]))
        else:
            production_company2.append(None)    
    except:
            production_company1.append(None)
            production_company2.append(None)    

print ('production_company1',len(production_company1))
```

```python
#editors 

editor=[]


for d in editors:
    try:
        editor.append(str(d[0]))
    except:
        editor.append(None) 

#composers 

composer=[]

for d in composers:
    try:
        composer.append(str(d[0]))
    except:
        composer.append(None) 
    
#cinematographers

cinematographer = []

for d in cinematographers:
    try:
        cinematographer.append(str(d[0]))
    except:
        cinematographer.append(None) 
    
#set_decorators

set_decorator=[]

for d in set_decorators:
    try:
        set_decorator.append(str(d[0]))
    except:
        set_decorator.append(None) 
    

#assistant_directors

assistant_director=[]

for d in assistant_directors:
   
    try:
        assistant_director.append(str(d[0]))
    except:
        assistant_director.append(None) 
    

#visual_effects

visual_effect=[]

for d in visual_effects:
    try:
        visual_effect.append(str(d[0]))
    except:
        visual_effect.append(None) 
```

```python

#Budget, Opening Weekend United States, Cumulative Worldwide Gross

Budget=[]
OWUS =[]
CWG = []

for d in box_office:
    try:
        Budget.append(int(d['Budget'].replace(' (estimated)','').replace('$','').replace(',','') ))
    except:
        Budget.append(None) 
    try:
        OWUS.append(d['Opening Weekend United States'])
    except:
        OWUS.append(None) 
    try:
        CWG.append(int(d['Cumulative Worldwide Gross'].replace('$','').replace(',','')))
    except:
        CWG.append(None) 
        
        
        
#genres
genre=[]

for d in genres:
   
    try:
        genre.append(str(d))
    except:
        genre.append(None) 
    

#genre1 = []
#genre2 = []
#genre3 = []

#for d in genres:
#    try:
#        genre1.append(d[0])
#    except:
#        genre1.append(None)
#    try:
#        genre2.append(d[1])
#    except:
#        genre2.append(None) 
#    try:
#        genre3.append(d[2])
#    except:
#        genre3.append(None) 
        
```
