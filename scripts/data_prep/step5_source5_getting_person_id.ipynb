```python
cast_info = pd.DataFrame()
cast_info['Name'] = data['assistant_directors']
```

```python
# This will input the the person name to the library and get the person ID
IDs = []
for n in data['assistant_directors']:
    try:
        cid = moviesDB.search_person(n)[0].getID()
        print(cid)
        IDs.append(cid)
    except:
        IDs.append(None)
cast_info['ID'] = IDs
```
