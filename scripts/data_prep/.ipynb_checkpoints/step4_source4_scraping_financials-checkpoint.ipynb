## After getting the movie info using IMDBpy we further Scrape Movie's financial info from the IMDB website to fill missing financial values in our dataset. (The website has updated financial information)

```python
import numpy as nm
import requests
import random
import re
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# Read the file from step3_source3
movies = pd.read_csv('../../../Files/second_IMDB_2009_2017_new.csv')
```

```python
movies.columns
```

```python
#Drop the null values
movies['imdbID'].dropna(axis = 0, how ='all', inplace = True)
```

```python
#Drop duplicates
movies['imdbID'].drop_duplicates(keep='first', inplace=True)
```

## start scraping

```python
# Fix the ID then loop over the ID's and scrape the financial data
ID = movies['imdbID']


id2=[]
b_txt2 = []
Budget2 = []
Opening2 = []
Gross2 = []
Cumulative2 = []

total = len(ID)
count=0
for j in ID:
    if not nm.isnan(j):
        if len(str(int(j)))<7:
            right = 7 - len(str(int(j)))
            i = (right * '0')+(str(int(j)))
        else:
            i = 'tt' + str(int(i))
    id2.append(i)
    print("This is ",i)
    print(str(count) + " / " +str(total))
    count = count+1
    try:
        if type(i) == str:
            link = 'https://www.imdb.com/title/{}/?openid.assoc_handle=imdb_us&openid.claimed_id=http%3A%2F%2Fspecs.openid.net%2Fauth%2F2.0%2Fidentifier_select&openid.identity=http%3A%2F%2Fspecs.openid.net%2Fauth%2F2.0%2Fidentifier_select&openid.mode=logout&openid.ns=http%3A%2F%2Fspecs.openid.net%2Fauth%2F2.0&openid.return_to=https%3A%2F%2Fwww.imdb.com%2Ftitle%2Ftt0172495%2F&'.format(i)
            print(link)
            response = requests.get(link)
            print(response.status_code)
            
            soup = BeautifulSoup(response.text, 'html.parser')
            ss = soup.find_all('div',class_="article")
            print('ok')
            try:
                c =0
                right1 = ''
                for x in ss:
                    c=c+1
                    #print(c,x)
                    if '<h3 class="subheading">Box Office</h3>' in str(x):
                        right1 = x.text

                s= right1.find('Box Office')
                e =right1.find('Company Credits')
                AOI = right1[s:e]

                b_txt2.append(link)
                try:
                    b = right1[right1.find('Budget'):right1.find('Opening')].split(':')[1].split(' ')[0].replace('$','').replace(',','').split('\n')[0]
                    Budget2.append(b)
                except:
                    Budget2.append(None)
                try:
                    o = right1[right1.find('Opening'):right1.find('Gross')].split(':')[1].split(' ')[1].replace('$','').replace(',','').replace(' ','').split('\n')[0]
                    Opening2.append(o)
                except:
                    Opening2.append(None)

                try:    
                    g = right1[right1.find('Gross'):right1.find('Cumulative')].split(':')[1].replace(' ','').replace('$','').replace(',','').replace('\n','')
                    Gross2.append(g)
                except:
                    Gross2.append(None)

                try:
                    cw = right1[right1.find('Cumulative'):right1.find('See more on IMDbPro')].split(':')[1].replace(' ','').replace('$','').replace(',','').replace('\n','')
                    Cumulative2.append(cw)
                except:
                    Cumulative2.append(None)

            except:
                print ("err",i)
                b_txt2.append(None)
                Budget2.append(None)
                Opening2.append(None)
                Gross2.append(None)
                Cumulative2.append(None)
    except:
            print ("error soup",i)
            b_txt2.append(None)
            Budget2.append(None)
            Opening2.append(None)
            Gross2.append(None)
            Cumulative2.append(None)
```

## Insert information into a dataframe

```python
nn = pd.DataFrame()
nn['ID'] = id2
print('index',len(nn['ID']))
print('b_txt2',b_txt2)
print('Budget',Budget2)
nn['txt'] = b_txt2

nn['Budget'] = Budget2
nn['Opening'] = Opening2
nn['Gross'] = Gross2
nn['Cumulative'] = Cumulative2
```
