```python
missing_movies = pd.read_csv('names_second_batch.csv')
```

```python
Title_ = []
Year_ = []
Rated_ = []
Released_ = []
Runtime_ = []
Genre_ = []
Director_ = []
Writer_ = []
Actors_ = []
Plot_ = []
Language_ = []
Country_ = []
Awards_ = []
Ratings_ = []
Metascore_ = []
imdbRating_ = []
imdbVotes_ = []
imdbID_ = []
Type_ = []
DVD_ = []
BoxOffice_ = []
Production_ = []
Website_ = []
Response_ = []


count = 0
for i in range(0,missing_movies.shape[0]):
#for id_ in missing_movies['name']:
    
    try:
    
        data = requests.get('http://www.omdbapi.com/?t={}+&y={}&apikey=b892805a'.format(missing_movies['name'][i],missing_movies['year'][i])).json()
        #data = requests.get('http://www.omdbapi.com/?i={}&apikey=559f8f7f'.format(id_)).json()

        try:
            Title_.append(data['Title'])
        except KeyError:
            Title_.append(None)

        try:
            Year_.append(data['Year'])
        except KeyError:
            Year_.append(None)

        try:
            Rated_.append(data['Rated'])
        except KeyError:
            Rated_.append(None)

        try:
            Released_.append(data['Released'])
        except KeyError:
            Released_.append(None)

        try:
            Runtime_.append(data['Runtime'])
        except KeyError:
            Runtime_.append(None)

        try:
            Genre_.append(data['Genre'])
        except KeyError:
            Genre_.append(None)

        try:
            Director_.append(data['Director'])
        except KeyError:
            Director_.append(None)

        try:
            Writer_.append(data['Writer'])
        except KeyError:
            Writer_.append(None)

        try:
            Actors_.append(data['Actors'])
        except KeyError:
            Actors_.append(None)

        try:
            Plot_.append(data['Plot'])
        except KeyError:
            Plot_.append(None)

        try:
            Language_.append(data['Language'])
        except KeyError:
            Language_.append(None)

        try:
            Country_.append(data['Country'])
        except KeyError:
            Country_.append(None)

        try:
            Awards_.append(data['Awards'])
        except KeyError:
            Awards_.append(None)

        try:
            Ratings_.append(data['Ratings'])
        except KeyError:
            Ratings_.append(None)

        try:
            Metascore_.append(data['Metascore'])
        except KeyError:
            Metascore_.append(None)

        try:
            imdbRating_.append(data['imdbRating'])
        except KeyError:
            imdbRating_.append(None)

        try:
            imdbVotes_.append(data['imdbVotes'])
        except KeyError:
            imdbVotes_.append(None)

        try:
            imdbID_.append(data['imdbID'])
        except KeyError:
            imdbID_.append(None)

        try:
            Type_.append(data['Type'])
        except KeyError:
            Type_.append(None)

        try:
            DVD_.append(data['DVD'])
        except KeyError:
            DVD_.append(None) 

        try:
            BoxOffice_.append(data['BoxOffice'])
        except KeyError:
            BoxOffice_.append(None) 

        try:
            Production_.append(data['Production'])
        except KeyError:
            Production_.append(None) 

        try:
            Website_.append(data['Website'])
        except KeyError:
            Website_.append(None) 

        try:
            Response_.append(data['Response'])
        except KeyError:
            Response_.append(None) 
            
    except:
        Title_.append(None)
        Year_.append(None)
        Rated_.append(None)
        Released_.append(None)
        Runtime_.append(None)
        Genre_.append(None)
        Director_.append(None)
        Writer_.append(None)
        Actors_.append(None)
        Plot_.append(None)
        Language_.append(None)
        Country_.append(None)
        Awards_.append(None)
        Ratings_.append(None)
        Metascore_.append(None)
        imdbRating_.append(None)
        imdbVotes_.append(None)
        imdbID_.append(None)
        Type_.append(None)
        DVD_.append(None)
        BoxOffice_.append(None)
        Production_.append(None)
        Website_.append(None)
        Response_.append(None)

    count= count +1
    print(count,end='')
```
