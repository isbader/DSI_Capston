```python
#Plot a heat map of the missing varibles
def missing_heat_map(DataFrame):
    fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize = (18, 6))
    sns.heatmap(DataFrame.isnull(), yticklabels=False, ax = ax, cbar=False, cmap='viridis')
    ax.set_title('Movies_')
    plt.show()
    
    for i in DataFrame:
        print(i,': %',int((DataFrame[i].isnull().sum()/len(DataFrame[i]))*100),\
              'With {} missing values'.format((DataFrame[i].isnull().sum())))
missing_heat_map(Movies_)
```

```python
import ast

imd = [] #stands for Internet Movie Database
rotten = [] #rotten tometo
metacritic = []

for i in df['Ratings']:
    if type(i) == str:
        a = ast.literal_eval(i)
            
        if len(a) > 2:
            imd.append(a[0])
            rotten.append(a[1])
            metacritic.append(a[2])
        elif len(a) > 1:
            imd.append(a[0])
            metacritic.append(a[1])
            rotten.append(None)
        elif len(a) == 1:
            imd.append(a[0])
            metacritic.append(None)
            rotten.append(None)    
        else:
            imd.append(None)
            rotten.append(None)
            metacritic.append(None)
    else:
        imd.append(None)
        rotten.append(None)
        metacritic.append(None)
        
```

```python
def names_to_count(df,col,win,nom):
    if 'Unnamed: 0' in nom.columns:
        del nom['Unnamed: 0']
    if 'Unnamed: 0' in win.columns:
        del win['Unnamed: 0'] 
    if 'id' in nom.columns:
        del nom['id']
    if 'id' in win.columns:
        del win['id']
    
        
    win.columns = (['2000','2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013',
           '2014', '2015', '2016', '2017', '2018', '2019', 'i'])
    nom.columns =  (['2000','2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013',
           '2014', '2015', '2016', '2017', '2018', '2019',  'i'])

    win.set_index('i')
    nom.set_index('i')

    tempact = pd.DataFrame()
    tempact['per'] = df[col]
    perwin = []
    pernom = []
    perw = []
    y = []
    name = []

    for index, row in df.iterrows():
        this_movie = str(row['fname'])
        this_year = str(row['fyear'])
        y.append(this_year)
        name.append(this_movie)
        try:
            w = list(win.loc[win['i'] == row[col]][this_year])[0]
            perwin.append(w)
        except:
            perwin.append(0)
        try:
            n = list(nom.loc[win['i'] == row[col]][this_year])[0]
            pernom.append(n)
        except:
            pernom.append(0)
        perw.append(perwin[-1] + pernom[-1]*.5)

    
    tempact['mname'] = name
    tempact['year'] = y
    tempact[col+'win'] = perwin
    tempact[col +'nom'] = pernom
    tempact[col +'w'] = perw
    return perw

```

```python
def fast_mode(df, key_cols, value_col):
    """ 
    Calculate a column mode, by group, ignoring null values. 

    Parameters
    ----------
    df : pandas.DataFrame
        DataFrame over which to calcualate the mode. 
    key_cols : list of str
        Columns to groupby for calculation of mode.
    value_col : str
        Column for which to calculate the mode. 

    Return
    ------ 
    pandas.DataFrame
        One row for the mode of value_col per key_cols group. If ties, 
        returns the one which is sorted first. 
    """
    return (df.groupby(key_cols + [value_col]).size() 
              .to_frame('counts').reset_index() 
              .sort_values('counts', ascending=False) 
              .drop_duplicates(subset=key_cols)).drop(columns='counts')
```

```python
def fixingcurr(column):
    
    import re 
    from forex_python.converter import CurrencyRates
    
    column = movies['{}'.format(column)]

    values1 = []
    values2 = []
    for i in column:
        if type(i) != float:
            res1 = "".join(re.split("[^a-zA-Z]*", i))
            res2 = "".join(re.split("[a-zA-Z]*", i))
            values1.append(res1)
            values2.append(res2)
        else:
            values1.append(None)
            values2.append(None)
    
    a = pd.DataFrame({'curr':values1,'budget':values2})
    
    def NoneConv(x):
        if len(x) == 0:
            return 'USD'
        else:
            return x
            
    a['curr'] = a['curr'].apply(lambda x: NoneConv(x) if type(x)==str else x)
        
    old_bud = []
    new_bud = []
    old_curr = []
    
    c = CurrencyRates()
    
    for i in range(0,a.shape[0]):
        if type(a['curr'][i]) == str:
            try:
                if a['curr'][i] == 'FRF':
                    new_bud.append(int(round(int(a['budget'][i])/ 7)))
                    old_curr.append(a['curr'][i])
                    old_bud.append(a['budget'][i])
                else:
                    b = c.convert(a['curr'][i], 'USD', int(a['budget'][i]))
                    old_bud.append(a['budget'][i])
                    new_bud.append(int(round(b)))
                    old_curr.append(a['curr'][i])
            except:
                new_bud.append(None)
                old_curr.append(a['curr'][i])
                old_bud.append(a['budget'][i])
        else:
            old_bud.append(a['budget'][i])
            new_bud.append(None)
            old_curr.append(a['curr'][i])
            
    return new_bud
        
```

```python
def fixingfinances(column):
    
    import re
    column = movies['{}'.format(column)]
    
    def isNaN(num):
        if num != num:
            return True
        else:
            False
            
    real_gross = []
    for i in column:
        if isNaN(i) != True:
            try:
                real_gross.append(int(i))
            except:
                real_gross.append(float(re.sub("[^0-9\.]","",i)))
        else:
            real_gross.append(None)
        
    return real_gross
    
```
